```python
#plotando um gráfico
sns.countplot(df['total_litres_of_pure_alcohol'])
plt.ylabel('consumo de alcool por país')
```

```python
#vamos diminuir o gráfico e analisar os países de maior consumo ou igual a 7 litros
consumo = df[df['total_litres_of_pure_alcohol']>=7]
```

```python
consumo.head()
```

```python
#criar um gráfico somente com os dados acima
sns.countplot(consumo['total_litres_of_pure_alcohol'])
plt.ylabel('% de consumo')
```
